## Step 1: Installing a Windows Subsystem for Linux

### Step 1a: Ubunt-Who?

Search for Windows PowerShell in your Windows search bar, then select Run as administrator.



When powershell opens type `wsl --install -d ubuntu` and sit tight until this process completes. When finished you will need to restart your computer.

### Step 1b: Let's try that again...

Restart your computer...

### Step 1c: Configure-Eight Ubuntu

After restarting (and hopefully successfully installing) launch Ubuntu. An easy way to do this is searching for Ubuntu in the search bar:<br><br>


Now after selecting Ubuntu you should see an Ubuntu terminal running. This means progress! If not and you run into an installation error there is likely an issue in your [BIOS configuration](https://ubuntu.com/tutorials/install-ubuntu-on-wsl2-on-windows-10#4-configure-ubuntu) (see the info box). Allow ubuntu to finish its initial setup. Once complete enter a username and password. Finally ensure things are up to date by running `sudo apt update` (entering your password when queried) and then `sudo apt upgrade` pressing Y when prompted.

## Step 2: Unlocking the Windows SSH Key-Maker

### Step 2a: Ensure you have SSH installed

Enter `sudo apt install openssh-client`. If installation is required proceed accordingly.

### Step 2b: Generate Your Cluster Skeleton Key

To generate your SSH key type `ssh-keygen -t rsa`. Press enter to proceed (unless of course your unhappy with the default key storage location). You will then be prompted for a password. At this point you have two options (for what it's worth I currently utilize option one for tscc login):
   1) **Omit a password by hitting enter again** - This will prevent you from having to enter your password when using your key (such as logging into TSCC) and can make your life easier, or at least make logging in a little quicker. Naturally though it also makes things less secure. <br> 
   2) **Enter a password** - This trades convenience for security. Specifically everywhere you attempt to connect via your ssh key you will be prompted for your password. The upside it that it protects you from a less than stellar individual accessing your connections with your public key. <br><br>
 <span style="color:cyan">**NOTE**:</span> You can also generate a new key-password pair each for each connection, so if/when you need to change between security and convenience you have the flexibility to do so.
